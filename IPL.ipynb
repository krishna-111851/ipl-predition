# LOGISTIC REGRISSION

```python
# now converting labeled data into one hot encoing (for clearity about one hot encoding goole it thenb see the images)
# transform no-numerical data to numerical data

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
trf
```

```python
# building the model

from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_prediction = pipe.predict(X_test)
```

```python
Y_prediction
```

```python
# for evulation part we have accuracy score

# accuracy score
from  sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_prediction)
```
